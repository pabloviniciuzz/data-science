# Trabalhando com o Íris Dataset como método de estudo

## Importando nossas bibliotecas

```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
# carregando o dataset

iris = datasets.load_iris()
```

## Visualizando algumas características da biblioteca

```python
iris.data
```

```python
# mapeamento da espécie de cada linha

iris.target
```

```python
# criando um dataframe

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# adicionando coluna de target

df['target'] = iris.target
```

```python
df.head()
```

```python
# verificando quais espécies existem para cada target

iris.target_names
```

```python
df['target'].value_counts()
```

```python
# definindo os nomes das espécies para cada target

dictNames = {
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
}
```

```python
# mapeando os target para substituir os valores 

df['target_names'] = df['target'].map(dictNames)
```

```python
# repare que os targets foram substituídos por espécies

df.head()
```

```python
df['target_names'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

## Iniciando a clusterização

```python
from sklearn.cluster import KMeans
```

```python
dfIris = df
```

```python
dfIris
```

```python
# retirando as colunas desnecessárias e deixando apenas os valores

flowers = dfIris.iloc[:, :4].values
```

```python
# armazenando o KMeans

km = KMeans(n_clusters=3)
```

```python
# criando o fit com os dados

km.fit(flowers)
```

```python
# descobrindo os centróides

km.cluster_centers_
```

```python
# encotrando os targets

yHat = km.fit_predict(flowers)
yHat
```

```python
# guardando os targets em nova coluna dos dados

dfIris['target_previsto'] = yHat
dfIris
```

```python
# visualizando a distribuição

sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=dfIris, hue='target_previsto')
plt.show()
```

```python
# comparando com a nossa target original e encontrando os erros nas interseções
# percebe-se que alguns acabam se misturando pois o KMeans tem essa dificuldade

sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=dfIris, hue='target')
plt.show()
```

```python
# visualizando com 4 clusters 

sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=dfIris, hue='target_previsto')
plt.show()
```

```python
# armazenando os targets em datasets de cada espécie

target_0 = dfIris[dfIris['target']==0]
target_1 = dfIris[dfIris['target']==1]
target_2 = dfIris[dfIris['target']==2]
```

```python
# como as frequencias são diferentes por espécie

sns.distplot(target_0['sepal length (cm)'], hist=False)
sns.distplot(target_1['sepal length (cm)'], hist=False)
sns.distplot(target_2['sepal length (cm)'], hist=False)
plt.show()
```

```python
flowers = dfIris.iloc[:, :4].values
```

```python
mscaler = MinMaxScaler()

# yHat previsto em cima de 'flowers' normalizado

flowers_min_max = mscaler.fit_transform(flowers)
km = KMeans(n_clusters=3)
yHatMM = km.fit_predict(flowers_min_max)
```

```python
# pegando o dataset original com o resultado no yHatMM nos originou, colocando-o em uma nova coluna

dfIris['target_min_max'] = yHatMM
```

```python
# fazendo a mesma coisa para o StandardScale

stdscaler = StandardScaler()

flowers_std = stdscaler.fit_transform(flowers)
km = KMeans(n_clusters=3)
yHatSTD = km.fit_predict(flowers_std)
```

```python
dfIris['target_std'] = yHatSTD
```

```python
# scatterplot do target original

sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target')
plt.show()
```

```python
# scatterplot do target_previsto

sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target_previsto')
plt.show()
```

```python
# scatterplot do sscale

sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target_std')
plt.show()
```

```python
# fazendo um plot com todos os gráficos

fig, axes = plt.subplots(1,4, figsize=(20, 4))
axes[0].set_title('Sem Normalização')
axes[1].set_title('Min Max Scaler')
axes[2].set_title('Standard Scaler')
axes[3].set_title('Original')

sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target_previsto', ax=axes[0])
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target_min_max', ax=axes[1])
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target_std', ax=axes[2])
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", data=dfIris, hue='target', ax=axes[3])
plt.show()
```

```python
# criando uma lista para armazenar todos os pesos

wcss = []
for i in range(1, 10):
  km = KMeans(n_clusters=i)
  km.fit(flowers)
  wcss.append(km.inertia_)
```

```python
# o 2 está quebrado pois temos um grupos distante dos outros, poderia clusterizar em 2
# porém depois ele infere clusterizar em 3

plt.plot(range(1,10), wcss)
plt.show()
```
