### Importando a biblioteca Neighbors do Sklearn para encontrar quem são as pessoas mais próximas de mim

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# primeiro precisa fitar, para ele conheça todos os dados que tem grupos de 7 pessoas

nbrs = NearestNeighbors(metric='euclidean', n_neighbors=7).fit(df_bank)
```

### Criando algumas variáveis

```python
age = 35
balance = 200
duration = 30
```

### Encontre para mim quem são os K-Vizinhos (7) dessas variáveis que criei

##### A resposta vai retornar dois valores: a distância e os índices (distances e indices)


```python
distances, indices = nbrs.kneighbors([[age, balance, duration]])
```

```python
distances
```

```python
indices[0]
```

```python
resultado = df_bank.iloc[indices[0]]
```

```python
# aqui estão os 7 que parecem comigo, sem estar normalizado

resultado
```
